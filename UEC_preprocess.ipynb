```python
import os
from PIL import Image
from sklearn.model_selection import train_test_split

# some images have multiple bounding boxes depending on the number of objects in the plate
# load bounding boxes:
def crop_and_sort_images(folder):
    folder = str(folder)
    base_folder = './data/'
    train_path = os.path.join(base_folder, 'train/', folder + '/')
    validation_path = os.path.join(base_folder, 'validation/', folder + '/')
    os.makedirs(train_path)
    os.makedirs(validation_path)

    boundaries_file = open('./' + folder + '/' + 'bb_info.txt')
    if boundaries_file.readline()[:3] != 'img':
        print ('WARNING: First line does not start with img')


    file_list = boundaries_file.readlines()
    boundaries_file.close()
    train, test = train_test_split(file_list, test_size = 0.3, random_state=2016)
    n = 0
    for line in train:        
        bounds = line.strip().split()
        image_in = './' + folder + '/' + str(bounds[0]) + '.jpg'
        object_count = 0
        image_output = train_path + bounds[0] + '_' + str(object_count) + '.jpg'
        while os.path.isfile(image_output):
            object_count += 1
            image_output = train_path + bounds[0] + '_' + str(object_count) + '.jpg'    

        box = [int(i) for i in bounds[1:]]       

        ## Crop the image with the same filename
        print ( 'Processed image:  ', image_output)
        image = Image.open(image_in)
        image.crop(box).save(image_output)
        n +=1
    print (n,' training images were cropped and copied from class ', folder)

    n = 0
    for line in test:        
        bounds = line.strip().split()
        image_in = './' + folder + '/' + str(bounds[0]) + '.jpg'
        object_count = 0
        image_output = validation_path + bounds[0] + '_' + str(object_count) + '.jpg'
        while os.path.isfile(image_output):
            object_count += 1
            image_output = validation_path + bounds[0] + '_' + str(object_count) + '.jpg'    


        box = [int(i) for i in bounds[1:]]       

        ## Crop the image with the same filename
        print ( 'Processed image:  ', image_output)
        image = Image.open(image_in)
        image.crop(box).save(image_output)
        n +=1
    print (n,' test images were cropped and copied from class ', folder)
```
